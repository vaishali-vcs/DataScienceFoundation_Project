# Purpose:

This code is used to generate the heatmap of suicide rate for every country in the Suicide dataset. 

**Data Preparation**

1. We have added the Lattitude and Longitude information for the countries in the Suicide dataset. This was done in excel
2. Dropped unnecessary columns except country, suicides_no, population and suicides100k pop.
2. the processed file is saved as cleaned_temperature_data.csv

**features of the cleaned dataset**

- country
- Lat
- Long
- suicides_no
- population
- suicides100k pop

**Libraries used**

- folium
- pandas

```python
import pandas as pd
import folium
from folium import plugins
from folium import map
from folium import features
```

```python
countries = pd.read_csv('countries_heatmap.csv',sep=',')
```

```python
countries['Lat'] = countries['Lat'].astype(float)
countries['Long'] = countries['Long'].astype(float)
```

```python
#Initialize the map and Plot suicides100k pop data using a heat map 
mapOutput = folium.Map([countries.Lat.mean(), countries.Long.mean()], height=500, zoom_start = 4, tiles='OpenStreetMap')   
mapOutput.add_child(plugins.HeatMap(zip(countries['Lat'],countries['Long'],countries['suicides100k pop']), radius = 10))
```
